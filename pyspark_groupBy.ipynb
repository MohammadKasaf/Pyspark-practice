```python
from pyspark.sql.functions import col

data = [
    (101,"Ali",25,"IT",50000,"Delhi","2023-01-10"),
    (102,"Rahul",28,"HR",45000,"Mumbai","2022-03-15"),
    (103,"Neha",30,"IT",60000,"Pune","2021-05-20"),
    (104,"Sara",None,"Finance",55000,"Delhi","2020-07-11"),
    (105,"John",35,"IT",None,"Bangalore","2019-09-30"),
    (106,"Amit",29,"HR",47000,None,"2022-06-18"),
    (107,"Priya",27,"Finance",52000,"Mumbai",None),
    (108,"Rahul",28,"HR",45000,"Mumbai","2022-03-15"),
    (109,"Neha",30,"IT",60000,"Pune","2021-05-20"),
    (110,"Arjun",31,"IT",-10000,"Delhi","2023-02-01")
]

schema="emp_id int,name String,age int,department String,salary int,city String,date String"

df=spark.createDataFrame(data,schema)
df.show()
df.printSchema()
df.count()
df.sort("salary",ascending=False).show()
df.select(col("name"),col("salary")).show()
df.filter(col("salary")>50000).show()
```

```python
from pyspark.sql.functions import avg

df_avg_salary=df.groupBy("department").agg(avg("salary").alias("average_salary"))
df_avg_salary.show()
```

```python
from pyspark.sql.functions import *


df_avg_salary=df.groupBy("department").agg(max("salary").alias("max_salary"))
df_avg_salary.show()

df_avg_salary=df.groupBy("department").agg(min("salary").alias("min_salary"))
df_avg_salary.show()
```

```python
from pyspark.sql.functions import *


df_avg_salary=df.groupBy("department").agg(count("salary").alias("total_employees"))
df_avg_salary.show()
```
