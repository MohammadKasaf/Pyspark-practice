```python
df=spark.read.csv("/Volumes/workspace/default/pyspark_files_volume/employees.csv",header=True,inferSchema=True)
df.show()
df.printSchema()
```

```python
df.select("emp_name","salary").show()
df.filter("salary>70000").show()
df.filter("department=='IT'").show()
df.filter(" age>=25 and age<=30 ").show()
df.filter("department='IT' or department='finance' ").show()
df.filter("emp_name like 'A%' ").show()
```

```python
from pyspark.sql.functions import when,col

df.withColumn("annual_salary",df.salary*12).show()
df.withColumn("bonus",df.salary*0.10).show()
df.withColumn("salary_category",
              
              when(col("salary")>80000,"high")
              .when(col("salary")>60000,"medium")
              .otherwise("low")
               
           
     ).show()
```

```python
from pyspark.sql.functions import upper,lower

df.withColumnRenamed("emp_name","employee_name").show()
df.withColumnRenamed("salary","employee_salary").show()
df.select(upper("emp_name")).show()
df.select(lower("city")).show()
```
