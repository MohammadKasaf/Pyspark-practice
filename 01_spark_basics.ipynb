1. What is Apache Spark?
Definition

Apache Spark is an open-source distributed data processing engine designed to process massive amounts of data quickly across multiple machines.

Simple Explanation

Imagine you have:

10 GB of data
100 GB of data
10 TB of data

A single computer may take hours to process this data.

Spark solves this problem by distributing the work across many computers and processing data in parallel.

Instead of:

One machine doing all work

Spark does:

Machine 1 -> Part A
Machine 2 -> Part B
Machine 3 -> Part C
Machine 4 -> Part D

All machines work simultaneously.

This is called:

Distributed Computing
Why Spark Was Created

Before Spark, Hadoop MapReduce was popular.

Problem with Hadoop:

Read Data
↓
Process
↓
Write to Disk
↓
Read Again
↓
Process Again
↓
Write Again

Too much disk I/O.

Very slow.

Spark introduced:

In-Memory Processing

Meaning:

Read Data Once
Keep in Memory
Process Multiple Times

Result:

10x to 100x Faster

for many workloads.

5. RDD vs DataFrame vs Dataset

Evolution:

RDD
 ↓
DataFrame
 ↓
Dataset
RDD

RDD means:

Resilient Distributed Dataset

Oldest Spark abstraction.

Characteristics:

Distributed
Fault Tolerant
Immutable

Example:

rdd = spark.sparkContext.parallelize([1,2,3,4])

Advantages:

Full Control

Disadvantages:

No Optimization
Slower
DataFrame

Most used abstraction today.

Structure:

Rows + Columns

Similar to:

SQL Table
Excel Sheet
Pandas DataFrame

Example:

df = spark.read.csv("employees.csv")

Advantages:

Faster
Optimized
Easy to Use
Dataset

Available mainly in Scala and Java.

Provides:

DataFrame Features
+
Type Safety

Rarely used in PySpark.

Comparison
Feature	RDD	DataFrame	Dataset
Schema	No	Yes	Yes
Optimization	No	Yes	Yes
Performance	Slow	Fast	Fast
Type Safety	No	No	Yes
PySpark Support	Yes	Yes	Limited
6. Transformations and Actions

This concept explains Spark's Lazy Evaluation.

Transformations

Transformation creates a new DataFrame.

No execution happens immediately.

Examples:

filter()
select()
groupBy()
join()
withColumn()
drop()

Example:

df2 = df.filter(df.salary > 50000)

Spark does not execute yet.

Actions

Actions trigger execution.

Examples:

show()
count()
collect()
first()
take()
write()

Example:

df2.show()

Now Spark starts processing.

Lazy Evaluation

Spark waits until an Action appears.

Example:

df = spark.read.csv("data.csv")

df = df.filter(...)

df = df.groupBy(...)

df = df.select(...)

Still no execution.

Execution starts only when:

df.show()

This allows Spark to optimize the entire workflow before running it.
